Write a Python program to implement a Stack class that has the following functions

Push Item
Pop Item
Print the Items in the stack
Size of Stack
The top item in the stack
Check if the stack is empty

```python

class stack():
    def __init__(self):
        self.store = []
        
    def collect(self):
        for i in range(4):
            self.item = input("enter value: ")
            self.store.append(self.item)
       
        for i in self.store:
            print(i)

    def remove(self):
        self.store.pop()
        for i in self.store:
            print(i)
             
s = stack()


while True:
    print("#1# push item")
    print("#2# pop item")
    print("#3# items")
    print("#4# length")
    print("#5# top item ")
    print("#6# exit")
    choice = int(input("enter your choice: "))
    if choice == 1:
        s.collect()
    if choice == 2:
        s.remove()
    if choice == 3:
        print("stored items",s.store)
    if choice == 4:
        print("length is",len(s.store))
    if choice == 5:
        print("the top item is",s.store[len(s.store)-1])
      
    if choice == 6:

        break


```
